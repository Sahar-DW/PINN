```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
#required parameters
length=10
k=0.89
temp_left=100
temp_right=200
total_time=10
```

```python
#discret of space
dx=0.1
x_vector=np.linspace(0,length,int(length/dx))
#discret of time
dt=0.1
t_vector=np.linspace(0,total_time,int(total_time/dt))
```

```python
#initaial and boundary condition
T=np.zeros([len(t_vector),len(x_vector)])
T[:,0]=temp_left
T[:,-1]=temp_right

#plot initial condition
plt.plot(x_vector,T[0,:])
plt.xlabel('X')
plt.ylabel('Temp')
plt.title('inital condition')
plt.show()
```

```python
#solve PDE
for t in range(1,len(t_vector)-1):
    for x in range(1,len(x_vector)-1):
        T[t+1,x]=(k*(dt/dx**2))*(T[t,x+1]-2*T[t,x]+T[t,x-1])+T[t,x]
```

```python
# Create the animation
fig, ax = plt.subplots()
line, = ax.plot(x_vector, T[0, :])
ax.set_ylim(0, 250)
ax.set_xlim(0, length)
ax.set_xlabel('X')
ax.set_ylabel('Temperature')
ax.set_title('Temperature Distribution Over Time')

def update(frame):
    line.set_ydata(T[frame, :])
    return line,

# Set interval to a lower value for faster updates
ani = FuncAnimation(fig, update, frames=len(t_vector), interval=20, blit=True)
plt.show()
```
